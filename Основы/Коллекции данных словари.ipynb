**Словари** - неупорядоченные коллекции произвольных объектов с доступом по ключу 

В отличие от других типов данных элемент в словаре Python состоит из двух частей:
* ключа (key), по которому мы ищем
* значения (value)
Они разделяются двоеточием: key:value

### Взаимодействие со словарем

````python
countries_dict = {
    "Россия":"Москва",
    "Белоруссия":"Минск",
    "Германия":"Берлин"
}

# Добавление
print(f"Изначальный словарь countries_dict = {countries_dict}")
countries_dict["Турция"] = "Анкара"
print(f"Добавлен элемент в словарь countries_dict = {countries_dict}")

# Безопасное добавление
# 1 Вариант
key = "Турция"
value = "Анкара"
if key not in countries_dict:
    countries_dict[key] = value
else:
    print(f'Ключ "{key}" уже есть в словаре!')

# 2 вариант 
key = "Англия"
value = "Лондон"  
if key not in countries_dict:
    countries_dict[key] = value
else:
    print(f'Ключ "{key}" уже есть в словаре!')
print(f"Добавлен элемент в словарь countries_dict = {countries_dict}")

# 3 Вариант

key = "Франция"
countries_dict.setdefault(key, "Париж")
print(countries_dict)


# Удаление
key = "Корея"
print(countries_dict.get(key, "Нет такого объекта"))

# key = "Англия"
# del countries_dict[key]
# print(f"Удален элемент из словаря countries_dict = {countries_dict}")

countries_dict_2 = {}
key = "Англия"
value = countries_dict.pop(key)

print(countries_dict)
print(key, value)
countries_dict_2[key] = value
print(countries_dict_2)

# Объединение словарей
countries_dict.update(countries_dict_2)
print(countries_dict)

# Количество элементов
print(f'Количество элементов = {len(countries_dict)}')

# Очистить словарь
countries_dict.clear()
print(countries_dict)
````

### Правила для словарей

* Словарь обычно создаётся функцией dict() или через {}
* Элемент словаря состоит из пары ключ:значение. Двоеточие между ними
обязательно
* Словарь записывается в фигурных скобках {}
* Элементы (пары ключ:значение) в словаре записываются через запятую
* Порядок элементов в словаре не соблюдается. Получить значение в словаре
по индексу (как в списке) нельзя
* Доступ к значению словаря — только по ключу
* Ключи в словаре должны быть уникальными, не могут повторяться
* Ключами могут быть только неизменяемые типы данных: числа, строки, кортежи
* Значениями могут быть любые типы данных, включая вложенные списки и словари
