## Правила для множества
* Множество создаётся функцией set() 
* Множество записывается в фигурных скобках {} 
* Элементы разделяются запятыми 
* Порядок элементов в множестве не соблюдается. 
Получить значение по индексу (как в списке) — нельзя 
* Множество обычно преобразовывают в список для удобства
* Элементы в множестве уникальны (не могут повторяться). 
Если преобразовать список в множество, повторяющиеся значения исчезнут 
* Элементы множества реализуют теорию множеств 

1. Объединение ко всем элементам первого множества прибавить все элементы второго множества и оставить уникальные элементы
   1. Пример: Множество А - 1, 2 ,3; Множество В - 3, 4, 5; Объединение: 1, 2, 3, 4, 5
2. Пересечение выделяем только общие пересекаемые элементы
   1. Пример Множество А - 1, 2, 3; Множество В - 3, 4, 5; Пересечение: 3
3. Разность из левого множества вычитаем правое
   1. Пример: Множество А - 1, 2 ,3; Множество В - 3, 4, 5; А - В = 1,2; В - В = 4, 5
4. Cимметричная разность объединяем множества, а затем вычитаем общие
   1.  Пример: Множество А - 1, 2 ,3; Множество В - 3, 4, 5; А ^ В = 1, 2, 4, 5
5.  Подмножество
    1.  Пример: Множество А - 1, 2 , 3, 4, 5; Множество В - 3, 4, 5; А подмножество В = нет, В подмножество А = да

````python
set_a = {1, 2, 3}
set_b = {3, 4, 5}
print(f'Пересечение обозначается: "&" set_a & set_b = {set_a & set_b}') 
print(f'Объединение обозначается: "|" set_a | set_b = {set_a | set_b}')
print(f'Разность обозначается: "-", порядок важен: set_a - set_b = {set_a - set_b}')
print(f'Разность обозначается: "-", порядок важен: set_b - set_a = {set_b - set_a}')
print(f'Симметричная разность обозначается: "^", set_b ^ set_a = {set_b ^ set_a}')
````
